1. Как зовут владельца, у недвижимости которого за отчетный период была наибольшая разница между максимальной и минимальной ценой?

```python
calendar['price'] = calendar['price'].apply(lambda x: str(x).replace('$', ''))
calendar['price'] = calendar['price'].apply(lambda x: str(x).replace(',', '')).astype(float)
```

```python
listings['price'] = listings['price'].apply(lambda x: str(x).replace('$', ''))
listings['price'] = listings['price'].apply(lambda x: str(x).replace(',', '')).astype(float)
listings.rename(columns={"price": "listing_price"}, inplace=True)
```

```python
tmp2 = pd.merge(listings, calendar.groupby('listing_id').agg({'price': 'max'}), left_on='id', right_on='listing_id', how='left')
tmp2.rename(columns={"price": "max_price"}, inplace=True)
tmp2 = pd.merge(tmp2, calendar.groupby('listing_id').agg({'price': 'min'}), left_on='id', right_on='listing_id', how='left')
tmp2.rename(columns={"price": "min_price"}, inplace=True)
```

```python
tmp2['price_range'] = tmp2["max_price"] - tmp2["min_price"]
tmp2 = tmp2.groupby(['host_id', 'host_name']).agg({'price_range': 'max'})
```

```python
tmp4 = pd.merge(reviews, listings, left_on='listing_id', right_on='id', how='left')
```

```python
tmp4 = tmp4[['host_id', 'host_name', 'reviewer_name', 'review_scores_accuracy', 'review_scores_communication', 'review_scores_location', 'review_scores_value']] 
```

```python
tmp5 = tmp4.loc[tmp4['host_name'] == tmp4['reviewer_name']].groupby('host_id').agg({'reviewer_name': 'count'})
```

```python
tmp4 = pd.merge(tmp4, tmp5, on='host_id')
```
